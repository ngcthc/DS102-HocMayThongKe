# Bài tập thực hành 1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Bài 1

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
X_b = np.c_[np.ones((100, 1)), X]
```

```python
plt.plot(X, y, "b.")
```

```python
X_b.shape
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
```

```python
theta_best = np.linalg.inv(X_b.T@X_b)@X_b.T@y
theta_best
```

```python
y_pred = X_new_b @ theta_best
y_pred
```

```python
plt.plot(X_b, y, "b.")
plt.plot(X_new_b, y_pred, "r-")
```

# Bài 3

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
X_b = np.c_[np.ones((100, 1)), X]

X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
```

```python
alpha = 0.02
theta_best = np.linalg.inv(X_b.T@X_b + alpha*np.eye(X_b.shape[1]))@X_b.T@y
theta_best
```

```python
y_pred_ridge = X_new_b @ theta_best
y_pred_ridge
```

```python
plt.plot(X_b, y, "b.")
plt.plot(X_new_b, y_pred_ridge, "r-")
```

# Bài 4

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
X_b = np.c_[np.ones((100, 1)), X]

X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
```

## So sánh Linear Regression tự code và thư viện

```python
#Tự code
theta_best = np.linalg.inv(X_b.T@X_b)@X_b.T@y
y_pred_lr = X_new_b @ theta_best
```

```python
#Thư viện
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_b, y)
y_pred_lr_sk = linear_model.predict(X_new_b)

plt.plot(X_new_b, y_pred_lr_sk, "b-")
plt.plot(X_new_b, y_pred_lr, "r-")
```

## So sánh Ridge Regression tự code và thư viện

```python
# Tự code
alpha = 0.02
theta_best = np.linalg.inv(X_b.T@X_b + alpha*np.eye(X_b.shape[1]))@X_b.T@y
y_pred_ridge = X_new_b @ theta_best
```

```python
# Thư viện
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=0.02)
ridge_reg.fit(X_b, y)

y_pred_ridge_sk = ridge_reg.predict(X_new_b)

plt.plot(X_new_b, y_pred_ridge_sk, "b-")
plt.plot(X_new_b, y_pred_ridge, "r-")
```

**Quan sát 2 biểu đồ ta nhận thấy:** kết quả khi chạy linear regression và ridge regression bằng 2 phương pháp tự cài đặt và dùng thư viên để cho kết quả như nhau.
